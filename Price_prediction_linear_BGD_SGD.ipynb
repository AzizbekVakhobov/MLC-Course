<h2>Splitting the Data</h2>

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['price'], axis=1)
y = df['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

<h2>Creating Pipeline</h2>

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#processing numeric values
num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')), #if there is a mising value i will fill it with a median robust t outliers
    ('scale', StandardScaler()) #normalize the data scale 
])

#processing categorical values
cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')), #filling missing values with the most frequent one
    ('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore')) # make a matrix of 1 and 0 for each unique value
])
```

```python
num_val = X_train.select_dtypes(include=[np.number]).columns
cat_val = X_train.select_dtypes(include=["object"]).columns

preprocessor = ColumnTransformer(transformers=[
    ('num', num_transformer, num_val),
    ('cat', cat_transformer, cat_val)
], remainder='drop')
```

<h2>Stochastic GD</h2>

```python
from sklearn.linear_model import SGDRegressor

complete_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', SGDRegressor(max_iter=1000, tol=1e-5, eta0=0.01, n_iter_no_change=100, random_state=42))
])
```

```python
complete_pipeline.fit(X_train,y_train)
```

<h4>Predictions</h4>

```python
prediction=complete_pipeline.predict(X_test)
prediction
```

<h4>Evaluation</h4>

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import math
```

```python
sns.scatterplot(x=prediction, y=y_test)
plt.title('Evaluation of our Model')
plt.xlabel('Predictions')
```

```python
print("Mean Absolute Error: ", mean_absolute_error(y_test, prediction))
print("Mean Squared Error: ", mean_squared_error(y_test, prediction))
print("Root Mean Squared Error RMSE: ", math.sqrt(mean_squared_error(y_test, prediction)))
```

```python
r2=r2_score(y_test,prediction)
r2
```

```python
comparison=y_test.copy()
comparison=pd.DataFrame(comparison)
comparison['Predicted Values']=np.round(prediction,2)
comparison
```

```python
sns.regplot(x="price", y="Predicted Values", data=comparison, line_kws={'color': 'red', 'linewidth': 2})

plt.show()
```

<h2>Batch GD</h2>

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num", Pipeline([
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler())
        ]), num_val),
        ("cat", Pipeline([
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("onehot", OneHotEncoder(handle_unknown="ignore"))
        ]), cat_val)
    ]
)

# Preprocess the data
X_train_processed = preprocessor.fit_transform(X_train)
y_train_processed = np.array(y_train).flatten()
```

```python
# Batch Gradient Descent Implementation
class BatchGradientDescent:
    def __init__(self, learning_rate=0.01, max_iter=1000, tol=1e-5):
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.tol = tol
        self.theta = None

    def fit(self, X, y):
        m, n = X.shape
        self.theta = np.zeros(n)  # Initialize weights
        for iteration in range(self.max_iter):
            predictions = X.dot(self.theta)
            errors = predictions - y
            gradient = (1 / m) * X.T.dot(errors)  # Compute gradient
            prev_theta = self.theta.copy()
            self.theta -= self.learning_rate * gradient  # Update weights
            # Check convergence
            if np.linalg.norm(self.theta - prev_theta, ord=2) < self.tol:
                print(f"Converged after {iteration + 1} iterations")
                break

    def predict(self, X):
        return X.dot(self.theta)
```

```python
# Train the model
bgd_model = BatchGradientDescent(learning_rate=0.01, max_iter=1000)
bgd_model.fit(X_train_processed, y_train_processed)
X_test_processed = preprocessor.transform(X_test)
```

```python
y_pred = bgd_model.predict(X_test_processed)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import math
print("Mean Absolute Error: ", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error: ", mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error RMSE: ", math.sqrt(mean_squared_error(y_test, y_pred)))
r2=r2_score(y_test,y_pred)
r2
```
