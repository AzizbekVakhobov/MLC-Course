<h2><b>About Dataset</b></h2>

The dataset provides a collection of game reviews from the Steam platform, making it suitable for natural language processing (NLP) and machine learning projects. The columns included are:

<br><b>id</b>: Unique identifier for the review. 
<br><b>app_id</b>: Game identifier on Steam.
<br><b>content</b>: Review text.
<br><b>author_id</b>: Identifier for the review's author.
<br><b>is_positive</b>: Label indicating whether the review is positive (1) or negative (0).

```python
import pandas as pd

df = pd.read_csv("output.csv")

def display_info(df):
    print("\nFirst 20 rows of the dataset:")
    display(df.head(20))  
    print(f"\nTotal Rows: {df.shape[0]}")
    print(f"Total Columns: {df.shape[1]}")


# Display information about the dataset
display_info(df)
```

```python
# Count occurrences of 'negative' and 'positive' in the 'is_positive' column
negative_count = df['is_positive'].str.contains('Negative', case=False, na=False).sum()
positive_count = df['is_positive'].str.contains('Positive', case=False, na=False).sum()

# Display results
print(f"Number of 'negative': {negative_count}")
print(f"Number of 'positive': {positive_count}")
```

**We see that number of positive reviews exeeds negative ones**

```python
# Group by 'app_id' and calculate counts for 'negative' and 'positive'
result = df.groupby('app_id')['is_positive'].apply(
    lambda group: pd.Series({
        'negative_count': group.str.contains('negative', case=False, na=False).sum(),
        'positive_count': group.str.contains('positive', case=False, na=False).sum()
    })
).reset_index()

# Display the result
print(result)
```

**Here we see number of negative and positive reviews for each distinct app_id**
