<h2>Python code</h2>

```python
class Library:
    def __init__(self, lstbooks):
        self.availablebooks=lstbooks

    def showBooks(self):
        print("Available books in the library: ")
        for book in self.availablebooks:
            print(book)

    def lendBook(self, requestedBook):
        if requestedBook in self.availablebooks:
            print("You have successfully borrowed book!")
        else:
            print("Sorry, it seems like it's not available at the moment:(")

    def addBook(self, returnedBook):
        self.availablebooks.append(returnedBook)
        print("Thanks for returning you borrowed book:)")





class Student:
    def requestBook(self):
        print("Enter the title of desired book: ")
        self.book = input()
        return self.book

    def returnBook(self):
        print("Enter the title of desired book you want to return: ")
        self.book = input()
        return self.book



books = [
    "The Pragmatic Programmer - Andrew Hunt & David Thomas",
    "Clean Code - Robert C. Martin",
    "Design Patterns - Erich Gamma, Richard Helm, Ralph Johnson, John Vlissides",
    "Introduction to Algorithms - Thomas H. Cormen, Charles E. Leiserson, Ronald L. Rivest, Clifford Stein",
    "Artificial Intelligence: A Modern Approach - Stuart Russell & Peter Norvig",
    "Python Crash Course - Eric Matthes",
    "Fluent Python - Luciano Ramalho",
    "Automate the Boring Stuff with Python - Al Sweigart",
    "Deep Learning - Ian Goodfellow, Yoshua Bengio, Aaron Courville",
    "Data Science from Scratch - Joel Grus"
]

def main():
    library = Library(books)
    student = Student()
    execution = True
    while execution:
        print("""
    =======LIBRARY MENU==========
    1. Display All Available Books
    2. Request A Book
    3. Return A Book
    4. Exit
            """)
        choice = int(input("Choose one option(eg 1 or 2): "))
        if choice == 1:
            library.showBooks()
        elif choice == 2:
            library.lendBook(student.requestBook())
        elif choice == 3:
            library.addBook(student.returnBook())
        elif choice == 4:
            execution = False

main()
```
