## Data Preparation

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.metrics import accuracy_score, classification_report, mean_absolute_error, mean_squared_error
```

```python
X_risk=df_risk_class.drop("Heart_Attack_Risk",axis=1)
y_risk=df_risk_class["Heart_Attack_Risk"]
```

```python
num_f = X_risk.select_dtypes(include=np.number).columns
cat_f = X_risk.select_dtypes(exclude=np.number).columns
```

```python
preprocessor_risk = ColumnTransformer([
    ('num', StandardScaler(), num_f),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_f)
])
```

```python
X_risk_train, X_risk_test, y_risk_train, y_risk_test = train_test_split(X_risk, y_risk, test_size=0.2, random_state=42)
```

## Pipeline Creation

```python
risk_pipeline = Pipeline([
    ('preprocessor', preprocessor_risk),
    ('knn', KNeighborsClassifier(n_neighbors=20))
])
```

```python
risk_pipeline.fit(X_risk_train, y_risk_train)
```

```python
y_risk_pred = risk_pipeline.predict(X_risk_test)
```

```python
k_values = range(1, 21)
cv_scores = []

for k in k_values:
    risk_pipeline = Pipeline([
        ("preprocessing", preprocessor_risk),  
        ("model", KNeighborsClassifier(n_neighbors=k))
    ])

    scores = cross_val_score(risk_pipeline, X_risk_train, y_risk_train, cv=5, scoring='accuracy')
    cv_scores.append(np.mean(scores))
plt.figure(figsize=(8, 6))
plt.plot(k_values, cv_scores, marker="o", linestyle="dashed", linewidth=2, markersize=6)
plt.title("K-Value vs Cross-Validated Accuracy")
plt.xlabel("K")
plt.ylabel("Cross-Validated Accuracy")
plt.xticks(k_values)
plt.show()

best_k = k_values[np.argmax(cv_scores)]
print(f"Best value of K: {best_k}")
```

## Evaluation and Testing

```python
print(f"Cross-validation Accuracy: {np.mean(cv_scores):.4f}")
```

```python
print(f"Test Accuracy: {accuracy_score(y_risk_test, y_risk_pred):.4f}")
print("Classification Report:\n", classification_report(y_risk_test, y_risk_pred))
```
